```python
def redondear_hora(dt, split):
    m = int(round(dt.time().minute/split, 0))*split
    if m == 60 or dt.time().hour == 4: 
        return time(dt.time().hour + 1, 0)   
    else:
        return time(dt.time().hour, m)   

n = 15
name_col = 'rg' + str(n)+'min'
data[name_col] = data['init_datetime'].apply(redondear_hora, split=n)
data.head(85)
```

```python
def sumar_time(tiempo1, tiempo2, op='sum'):
    # Convertir los objetos time a objetos datetime
    dt1 = datetime.combine(datetime.today(), tiempo1)
    dt2 = datetime.combine(datetime.today(), tiempo2)

    # Sumar los objetos datetime
    if op == 'sum':
        resultado_dt = dt1 + timedelta(hours=dt2.hour, minutes=dt2.minute, seconds=dt2.second)
    elif op == 'res':
        resultado_dt = dt2 - timedelta(hours=dt1.hour, minutes=dt1.minute, seconds=dt1.second)
        
    # Convertir el resultado de datetime a time
    resultado_time = resultado_dt.time()

    # Imprimir el resultado
    return resultado_time
```
